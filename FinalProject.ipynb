```python
import json 
import requests 
import pandas as pd 
import csv
import matplotlib.pyplot as plt
import scipy.stats as st
from config import api_key
pd.set_option('display.max_rows', 9999)
```

```python
#This is how we get our key working as well as our url 
name= []
rating = []
type1 =[]
lat = []
long = []

# NW - NE - SE - SW
#lon2 = ["38.944993","38.925193", "38.900244", "38.889729"]
#lat2 = ['-77.078362','-76.985428',"-77.037211", "-76.989957"]
headers = {'Authorization': 'Bearer %s' % api_key}
url='https://api.yelp.com/v3/businesses/search'

for offset in range(0, 1000, 50):
    params = {
                'limit': 50, 
                "latitude":"38.944993",
                "longitude": "-77.078362",
                'term': "restaurant",
                'offset': offset, 
                
     
        }
    req = requests.get(url, params=params, headers=headers)
    data = json.loads(req.text)


    for i in range(50):
     
        try:
            type1.append(data["businesses"][i]["categories"][0]["title"])
            name.append(data["businesses"][i]["name"])
            rating.append(data["businesses"][i]["rating"])
            lat.append(data["businesses"][i]["coordinates"]["latitude"])
            long.append(data["businesses"][i]["coordinates"]["longitude"])
        except:
            print(f'There is nothing on page {offset} in row {i}')
    
```

```python
#NW DATAFRAME
NW = pd.DataFrame([name, type1, rating, long, lat]).T
NW

dictionary = {'name': name,
              "category": type1, 
              "latitude": lat, 
              "longitude":long, 
              'rating': rating,
              'region': "Northwest"}



NW = pd.DataFrame(dictionary)
NW
```
