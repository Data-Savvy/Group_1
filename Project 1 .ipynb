```python
import json 
import requests 
import pandas as pd 
import csv
import matplotlib.pyplot as plt
import scipy.stats as st
from config import api_key
```

```python
#This is how we get our key working as well as our url 
headers = {'Authorization': 'Bearer %s' % api_key}
url='https://api.yelp.com/v3/businesses/search'
params={'term':'restaurant', 'location':'Washigton DC'}

#Request 
req = requests.get(url, params=params, headers=headers)

#Call 
data = json.loads(req.text)
print(json.dumps(data, indent=4))

    
```

```python
businesses = data["businesses"]
for business in businesses:
    print("Name:", business["name"])
    print("Rating:", business["rating"])
    print("Type:", " ".join(business["categories"][0]["title"]))
    print("Longitude:", " "+str(business["coordinates"]["latitude"]))
    print("Latitude:", " "+str(business["coordinates"]["longitude"]))
    print("\n")


#USE LONGITUDE LATITUDE FOR GMAP! THANKS JURGEN! 
```
