```python
import requests
import json
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter


pd.set_option('display.max_rows', 9999)
```

```python
#Southwest Ethnicity Data
zip_20024_Hisp = pd.read_csv('DC_Health_Data/Hispanic_Latino_Population_by_Race_Zip_Code_20024.csv')
zip_20024_Non = pd.read_csv('DC_Health_Data/Non_Hispanic_Latino_by_Race_Zip_Code_20024.csv')

zip_20032_Hisp = pd.read_csv('DC_Health_Data/Hispanic_Latino_Population_by_Race_Zip_Code_20032.csv')
zip_20032_Non = pd.read_csv('DC_Health_Data/Non_Hispanic_Latino_by_Race_Zip_Code_20032.csv')

#Southeast Data
zip_20003_Hisp = pd.read_csv('DC_Health_Data/Hispanic_Latino_Population_by_Race_Zip_Code_20003.csv')
zip_20003_Non = pd.read_csv('DC_Health_Data/Non_Hispanic_Latino_by_Race_Zip_Code_20003.csv')

Ward_8_Hisp = pd.read_csv('DC_Health_Data/Hispanic_Latino_Population_by_Race_Ward_Ward_8.csv')
Ward_8_Non = pd.read_csv('DC_Health_Data/Non_Hispanic_Latino_by_Race_Ward_Ward_8.csv')

#Northeast Data
Ward_5_Hisp = pd.read_csv('DC_Health_Data/Hispanic_Latino_Population_by_Race_Ward_Ward_5.csv')
Ward_5_Non = pd.read_csv('DC_Health_Data/Non_Hispanic_Latino_by_Race_Ward_Ward_5.csv')

#Northwest Data
Ward_3_Hisp = pd.read_csv('DC_Health_Data/Hispanic_Latino_Population_by_Race_Ward_Ward_3.csv')
Ward_3_Non = pd.read_csv('DC_Health_Data/Non_Hispanic_Latino_by_Race_Ward_Ward_3.csv')

Ward_2_Hisp = pd.read_csv('DC_Health_Data/Hispanic_Latino_Population_by_Race_Ward_Ward_2.csv')
Ward_2_Non = pd.read_csv('DC_Health_Data/Non_Hispanic_Latino_by_Race_Ward_Ward_2.csv')

Ward_4_Hisp = pd.read_csv('DC_Health_Data/Hispanic_Latino_Population_by_Race_Ward_Ward_4.csv')
Ward_4_Non = pd.read_csv('DC_Health_Data/Non_Hispanic_Latino_by_Race_Ward_Ward_4.csv')

Ward_1_Hisp = pd.read_csv('DC_Health_Data/Hispanic_Latino_Population_by_Race_Ward_Ward_1.csv')
Ward_1_Non = pd.read_csv('DC_Health_Data/Non_Hispanic_Latino_by_Race_Ward_Ward_1.csv')
```

```python
#Southwest Data
zip_20024_Hispanic_sum = zip_20024_Hisp["Hispanic/Latino Population by Race"].sum()
zip_20024_Non = zip_20024_Non.append({'Category':"Hispanic",'Non-Hispanic/Latino by Race':zip_20024_Hispanic_sum},ignore_index=True)

zip_20024_Non = zip_20024_Non.rename(columns = { "Category":"Race","Non-Hispanic/Latino by Race":"Population"})
sw0_pop = zip_20024_Non.drop([5, 6]).reset_index(drop=True)


zip_20032_Hispanic_sum = zip_20032_Hisp["Hispanic/Latino Population by Race"].sum()
zip_20032_Non = zip_20032_Non.append({'Category':"Hispanic",'Non-Hispanic/Latino by Race':zip_20032_Hispanic_sum},ignore_index=True)

zip_20032_Non = zip_20032_Non.rename(columns = { "Category":"Race","Non-Hispanic/Latino by Race":"Population"})
sw1_pop = zip_20032_Non.drop([5, 6]).reset_index(drop=True)


psw_po = pd.DataFrame({"Race": sw0_pop["Race"],"Population": sw0_pop["Population"].add(sw1_pop["Population"])})
sw_pop=sw_pop.set_index("Race")
sw_pop
```

```python
sw_plot = sw_pop.plot(kind='bar',title = "Southwest DC Population By Race",rot=0,
                      figsize=(13,7),grid=True)

sw_plot.set_xlabel("Race")
sw_plot.set_ylabel("Population")

plt.ylim(0, 45000)
plt.yticks(np.arange(0,45000, 2000.0))

plt.tight_layout()
plt.savefig("DC Ethnic Plots/Southwest DC Population By Race.png")
```

```python
#Southeast Ethnicity Data
zip_20003_Hispanic_sum = zip_20003_Hisp["Hispanic/Latino Population by Race"].sum()
zip_20003_Non = zip_20003_Non.append({'Category':"Hispanic",'Non-Hispanic/Latino by Race':zip_20003_Hispanic_sum},ignore_index=True)

zip_20003_Non = zip_20003_Non.rename(columns = { "Category":"Race","Non-Hispanic/Latino by Race":"Population"})
se0_pop = zip_20003_Non.drop([5, 6]).reset_index(drop=True)

W8_Hispanic_sum = Ward_8_Hisp["Hispanic/Latino Population by Race"].sum()
Ward_8_Non = Ward_8_Non.append({'Category':"Hispanic",'Non-Hispanic/Latino by Race':W8_Hispanic_sum},ignore_index=True)

Ward_8_Non = Ward_8_Non.rename(columns = { "Category":"Race","Non-Hispanic/Latino by Race":"Population"})
se1_pop = Ward_8_Non.drop([5, 6]).reset_index(drop=True)

se_pop = pd.DataFrame({"Race": se0_pop["Race"],"Population": se0_pop["Population"].add(se1_pop["Population"])})
se_pop=se_pop.set_index("Race")
se_pop
```

```python
se_plot = se_pop.plot(kind='bar',title = "Southeast DC Population By Race",rot=0,
                      figsize=(14,7),grid=True)

se_plot.set_xlabel("Race")
se_plot.set_ylabel("Population")

plt.ylim(0, 86000)
plt.yticks(np.arange(0,86000, 3000.0))

plt.tight_layout()
plt.savefig("DC Ethnic Plots/Southeast DC Population By Race.png")
```

```python
#Northeast Ethnicity Data

W5_Hispanic_sum = Ward_5_Hisp["Hispanic/Latino Population by Race"].sum()
Ward_5_Non = Ward_5_Non.append({'Category':"Hispanic",'Non-Hispanic/Latino by Race':W5_Hispanic_sum},ignore_index=True)

Ward_5_Non = Ward_5_Non.rename(columns = { "Category":"Race","Non-Hispanic/Latino by Race":"Population"})
ne_pop = Ward_5_Non.drop([5, 6]).reset_index(drop=True)
ne_pop=ne_pop.set_index("Race")

ne_pop
```

```python
#Northwest Ethnicity Data
W3_Hispanic_sum = Ward_3_Hisp["Hispanic/Latino Population by Race"].sum()
Ward_3_Non = Ward_3_Non.append({'Category':"Hispanic",'Non-Hispanic/Latino by Race':W3_Hispanic_sum},ignore_index=True)

Ward_3_Non = Ward_3_Non.rename(columns = { "Category":"Race","Non-Hispanic/Latino by Race":"Population"})
nw3_pop = Ward_3_Non.drop([5, 6]).reset_index(drop=True)

W2_Hispanic_sum = Ward_2_Hisp["Hispanic/Latino Population by Race"].sum()
Ward_2_Non = Ward_2_Non.append({'Category':"Hispanic",'Non-Hispanic/Latino by Race':W2_Hispanic_sum},ignore_index=True)

Ward_2_Non = Ward_2_Non.rename(columns = { "Category":"Race","Non-Hispanic/Latino by Race":"Population"})
nw2_pop = Ward_2_Non.drop([5, 6]).reset_index(drop=True)


W1_Hispanic_sum = Ward_1_Hisp["Hispanic/Latino Population by Race"].sum()
Ward_1_Non = Ward_1_Non.append({'Category':"Hispanic",'Non-Hispanic/Latino by Race':W1_Hispanic_sum},ignore_index=True)

Ward_1_Non = Ward_1_Non.rename(columns = { "Category":"Race","Non-Hispanic/Latino by Race":"Population"})
nw1_pop = Ward_1_Non.drop([5, 6]).reset_index(drop=True)


W4_Hispanic_sum = Ward_4_Hisp["Hispanic/Latino Population by Race"].sum()
Ward_4_Non = Ward_4_Non.append({'Category':"Hispanic",'Non-Hispanic/Latino by Race':W4_Hispanic_sum},ignore_index=True)

Ward_4_Non = Ward_4_Non.rename(columns = { "Category":"Race","Non-Hispanic/Latino by Race":"Population"})
nw4_pop = Ward_4_Non.drop([5, 6]).reset_index(drop=True)


nw_pop = pd.DataFrame({"Race": nw1_pop["Race"],
                       "Population": nw1_pop["Population"].add(nw2_pop["Population"].add(nw3_pop["Population"].add(nw4_pop["Population"])))})
nw_pop=nw_pop.set_index("Race")
nw_pop
```
